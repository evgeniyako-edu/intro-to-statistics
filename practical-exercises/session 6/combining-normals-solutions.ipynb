# Combinations of normally distributed variables

In this notebook, you'll continue working with the Normal distribution. This time, you'll explore what will happen if you add up sevral normally ditributed variables.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

N = 1000
```

## From standard to any normal distribution

Imagine that you can only sample from the *standard* normal distribution ($\mu=0, \sigma=1$). How would you get a normally distributed random variable with some other mean and std?

First, sample N samples from the standard normal distribution.
You can do so, for example, with the help of the [np.random.randn()](https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randn.html) or [np.random.normal()](https://docs.scipy.org/doc//numpy-1.10.4/reference/generated/numpy.random.normal.html) functions.

Then, construct samples from the normal distribution with mean 10 and standard deviation 5.

```
# Your code here

X = np.random.randn(N)
Y = 5*X + 10
```

Plot histograms of the two samples to make sure that your approach works properly.

```
# Your code here

f = plt.figure(figsize=(10,3))
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3))
sns.histplot(X, ax=ax1)
ax1.set_title('$\mu=0, \sigma=1$')
sns.histplot(Y, ax=ax2)
ax2.set_title('$\mu=10, \sigma=5$')
```

## Linear combination of normally distributed variables

Draw three samples of size N from the normal distribution, each time with different parameter values:

X1 ~ N(10, 2)

X2 ~ N(3, 1)

X3 ~ N(5, 3)

For sampling from normal distribution with given parameter values, you can use, for example, [np.random.normal()](https://docs.scipy.org/doc//numpy-1.10.4/reference/generated/numpy.random.normal.html) function.

```
# Your code here

X1 = np.random.normal(10, 5, N)
X2 = np.random.normal(3, 3, N)
X3 = np.random.normal(5, 2, N)
```

Now, construct a new  variable Y which is a linear combination of X1, X2 and X3:

$Y = X1 + 2*X2 + 3*X3$



```
# Your code here

Y = X1 + 2*X2 + 3*X3
```

What are the mean, variance and standard deviation of $Y$?

```
# Your code here

m = 10 + 2*3 + 3*5
var = 1*(5**2) + (2**2)*(3**2) + (3**2)*(2**2)
std = np.sqrt(var)

print('mean: ', m, '\nvar: ', var, '\nstd: ', std)
```

Then, plot it's histogram of Y. 

What does it look like? Which distribution does Y seem to follow? 

```
# Your code here

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3))

sns.histplot(X1, ax=ax1)
_ = ax1.set_title('X1 ~ N(10, 5)')

sns.histplot(X2, ax=ax2)
_ = ax2.set_title('X2 ~ N(3, 3)')

sns.histplot(X3, ax=ax3)
_ = ax3.set_title('X3 ~ N(5, 2)')
```
