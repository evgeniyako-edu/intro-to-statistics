# Normal Distribution

In this notebook, you'll familiarize yourself with one of the most important continuous distributions - the Normal distribution.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

As you already know, a normally distributed random variable has two parameters, namely $\mu$ and $\sigma$.

The PDF of the Normal distribution is as follows:

$p(x) = \frac{1}{{\sigma\sqrt{2\pi}}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$

Let's get a better understanding how different values of the parameters $\mu$ and $\sigma$ influence the shape of the distribution.

First, let's get $n = 1000$ samples from the **standard** normal distribution, for which $\mu$ = 0 and $\sigma$ = 1. You can do so with the help of the [np.random.normal()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) function.

```
# Your code here

n = 1000
```

Now, compute the values of the probability density function of the normal destribution over the same range of values as you sampled.

Remember that for now, $\mu = 0$ and $\sigma = 1$.

```
def pdf_normal(x, mu, s):
  return (1./(s*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)/s)**2)
```

Now, plot the histogram of the sampled data. Then, add the PDF you've computed at the previous step to see how similar the theoretical (PDF) and empirical (histogram of the sampled data) distributions are.

```
# Your code here

mu = 0
s = 1

data = np.random.normal(mu, s, n)

x = np.linspace(min(data)-1, max(data)+1, 1000)
y = pdf_normal(x, mu, s)

sns.histplot(data, stat='density', label='data')
plt.plot(x, y, 'red', label='pdf')

_ = plt.legend()
```

Now change the values of the parameters of the distribution and repeat all the previous steps. 

Investigate how the theoretical shape of the distribution (the PDF) and the smaples change depending on the parameter values.

```
n = 1000
mu = 10
s = 1
data = np.random.normal(mu, s, n)

x = np.linspace(min(data)-1, max(data)+1, 1000)
y = pdf_normal(x, mu, s)

sns.histplot(data, stat='density', label='data')
plt.plot(x, y, 'red', label='pdf')

_ = plt.legend()
```

```
n = 1000
mu = 0
s = 50
data = np.random.normal(mu, s, n)

x = np.linspace(min(data)-1, max(data)+1, 1000)
y = pdf_normal(x, mu, s)

sns.histplot(data, stat='density', label='data')
plt.plot(x, y, 'red', label='pdf')

_ = plt.legend()
```
