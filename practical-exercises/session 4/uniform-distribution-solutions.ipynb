## Uniform distribution

### CDF and PDF

Do you remember te formula for the CDF and PDF of the uniform distribution?

First, define a function uniform_cdf() that returns the the values of the CDF of the uniform distribution at the given points x for the given values of the parameters $a$ and $b$.

```
# Your code here

def uniform_cdf(x, a, b):
  result = []
  for xx in x:
    if xx < a:
      result.append(0)
    elif xx <= b:
      result.append((xx-a)/(b-a))
    else:
      result.append(1)
  return result
```

Now, define a function uniform_pdf() that returns the the values of the PDF of the uniform distribution at the given points x for the given values of the parameters $a$ and $b$.

```
# Your code here

def uniform_pdf(x, a, b):
  result = []
  for xx in x:
    if xx < a:
      result.append(0)
    elif xx <= b:
      result.append(1/(b-a))
    else:
      result.append(0)
  return result
```

Now, visualize both CDF and PDF for some values of the parameters $a$ and $b$ of your choice.

```
# Your code here

n=1000

x=np.linspace(0, 6, n)

a = 1
```

```
for b in range(2, 6):
  label = 'a = ' + str(a) + ', b = ' + str(b)
  plt.plot(x, uniform_cdf(x, a, b), label=label)

_ = plt.legend(loc='lower right')
_ = plt.title('CDF of the Uniform Distribution')
```

```
for b in range(2, 6):
  label = 'a = ' + str(a) + ', b = ' + str(b)
  plt.plot(x, uniform_pdf(x, a, b), label=label)

_ = plt.legend(loc='upper right')
_ = plt.title('PDF of the Uniform Distribution')
```

### Mean and variance

Compute the theoretical mean and variance of the uniform random variable with the parameters $a$ and $b$ that you specified.

```
# Your code here

a = 0
b = 1

m = (a+b)/2
var = ((a-b)**2)/12

print('Theoretical mean: ', m)
print('Theoretical variance: ', var)
```

Now, compute the *sample* mean and *sample* variance based on the samples you've drawn. 

Are they close to their theoretical values?

```
n = 100

for i in range(6):
  print('SAMPLE ', i)
  data = np.random.random(n)
  m = np.mean(data)
  var = np.var(data)

  print('Sample mean: ', np.round(m, 3))
  print('Sample variance: ', np.round(var, 3))
```
