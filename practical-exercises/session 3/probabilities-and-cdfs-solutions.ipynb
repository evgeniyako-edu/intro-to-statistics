## Probability to CDF

Define a function that converts probability function of a discrete random variable X into a cumulative distribution function (CDF).

The input to your function is an array of probabilities [p0, p1, ..., pn] where pi = P(X = i). 

For example, probabilities array [0.2, 0.3, 0.5] corresponds to a random variable X that takes values 0, 1 and 2 with probabilities 0.2, 0.3 and 0.5 respectively. The output CDF for such a distribution should be [0.2, 0.5, 1].

```
# Your code here

def probabilities_to_cdf(probabilities):
  cdf = np.cumsum(probabilities)
  return cdf
```

## CDF to Probability

Define a function that converts a cumulative distribution function(CDF) of a discrete random variable into its probability function.

The input to your function is an array of values of CDF [F(0), F(1), ..., F(n)].

For example, array [0.2, 0.5, 1] represents the follwoing CDF: F(0) = 0.2, F(1) = 0.5, F(2) = 1.

The output probabilities array for such a distribution should be [0.2, 0.3, 0.5].

```
# Your code here

def cdf_to_probabilities(cdf):
  probabilities = []

  probabilities.append(cdf[0])

  for i in range(1,len(cdf)):
    probabilities.append(cdf[i]-cdf[i-1])

  return probabilities
```

## Binomial distribution

Define a function that generates the values of the probability function for the Binomial distribution with parameters n and p.

For example, binomial(2, 0.5) should result in [0.25, 0.5, 0.25], meaning that P(X=0) = P(X=2) = 0.25 and P(X=1) = 0.5.

```
# Your code here

def binomial(n, p):
  probabilities = []
  for k in range(n+1):
    probabilities.append(scipy.special.binom(n, k)*(p**k)*((1-p)**(n-k)))
  return probabilities
```

Check that your function works correctly: run it for some values of the parameters for which you can easily compute probabilities by hand:

## Uniform distribution

Define a function that computes the CDF of a uniform discrete distribution over n points.

For example, CDF [0.33, 0.66, 0.99] corresponds to a uniform discrete distribution over three points.


```
# Your code here

def cdf_uniform_discrete(n):
  cdf = [i/n for i in range(1,n+1)]
  return cdf
```

Now, define a function that computes CDF for the continuous uniform distribution between a and b at the given points x.


```
# Your code here

def cdf_uniform_continuous(x, a, b):
  cdf = (x-a)/(b-a)
  return cdf
```
